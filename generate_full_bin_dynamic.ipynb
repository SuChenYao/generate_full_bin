```python
import random
import math
import os
#variable setting
input_w = 6
input_h = 10
output_slice_number = 12
file_name = "new_slices.txt"
#generate full data
def generate_full_bin(bin_w,bin_h,slice_number):
  # check size and slice number
  if slice_number > bin_w * bin_h:
    print("couldn't cut out these many slices, max slice number is: ",bin_w * bin_h)
    return
  #width
  #determine how many slice on x side
  #randomly choose 2 integer in 1 to width
  x_slice_number = random.randint(1,min(slice_number,bin_w)) 
  if x_slice_number * bin_h < slice_number:
    x_slice_number = math.ceil(slice_number / bin_h)
  print("x_slice_number:",x_slice_number)

  x_cut_point = []
  x_cut_point = random.sample(range(1,bin_w),x_slice_number-1)
  x_cut_point.sort()
  print("x_cut_point:",x_cut_point)

  width_data = []
  if x_slice_number == 1:
      width_data = [bin_w *100] * 1
  else:
    for now in range(len(x_cut_point)):
      if now == 0  :
        width_data.append(x_cut_point[now]*100)
      else:
        width_data.append(x_cut_point[now]*100 - x_cut_point[now-1]*100)
    width_data.append(bin_w*100 - x_cut_point[len(x_cut_point)-1]*100)
  print("width_data:",width_data)

  #height
  #generate x_cut_number+1 set of y cutpoint total slice_number - 3 cutpoint(already x slice 3 slice)
  #every cut will increase 1 slice
  slice_remain = slice_number - x_slice_number
  y_slice_each = [1] * (x_slice_number)
  
  print("y_slice_each:",y_slice_each)
  print("slice_remain:",slice_remain)
  
  while(slice_remain > 0 and y_slice_each != [bin_h] * (x_slice_number)): # repeat until enough slices
    if x_slice_number == 1:
      y_slice_each[0] = min(slice_remain,bin_h)
    else:
      for each_y in range(len(y_slice_each)):  
        if each_y == len(y_slice_each) - 1 :
          #most bin_h slice per y,but already get some, most add (bin_h-1)
          add_slice = min(slice_remain, bin_h - y_slice_each[each_y]) 
          y_slice_each[each_y] += add_slice
        elif slice_remain == 0:
          break
        else:

          add_slice = random.randint(0, min(slice_remain, bin_h - y_slice_each[each_y]))
          y_slice_each[each_y] += add_slice

        slice_remain = slice_remain - add_slice

  print("y_slice_each after:",y_slice_each)
  print("slice_remain after:",slice_remain)

  #determine y cut point
  y_cut_point = []
  for count in range(x_slice_number):
    y_cut_point.append([])
    #each collumn has x slices, x-1 cut point
    y_cut_point[count] = random.sample(range(1,bin_h),y_slice_each[count]-1) 
    y_cut_point[count].sort()
   
  print("y_cut point:",y_cut_point)

  #generate collumn height data
  height_data = []
  for count in range(x_slice_number):
    height_data.append([])
    if y_slice_each[count] == 1:
      height_data[count] = [bin_h *100] * 1
    else:
      for now in range(len(y_cut_point[count])):
        if now == 0  :
          height_data[count].append(y_cut_point[count][now]*100)
        else:
          height_data[count].append(y_cut_point[count][now]*100 - y_cut_point[count][now-1]*100)
      height_data[count].append(bin_h*100 - y_cut_point[count][len(y_cut_point[count])-1]*100)
  print("height_data:",height_data)



  #combine width and height data to generate all 12 slice data string
  slice_data_string = ""
  for x_count in range(x_slice_number):
    for y_count in range(y_slice_each[x_count]):
      slice_data_string = slice_data_string + str(width_data[x_count]) + " " + str(height_data[x_count][y_count]) + "\n"

  return slice_data_string
generated_string = generate_full_bin(input_w, input_h, output_slice_number)
print("full bin:\n",generated_string)

#delete output file last time
try:
    os.remove(file_name) 
except OSError as e:
    print(e)
else:
    print("File is deleted successfully")

with open(file_name, 'a') as f:
  f.write(generated_string)
```
